```python
import sys

class Employee:
    def __init__(self):
        self.num = 0
        self.salary = 0
        self.name = ''
        self.next = None

def findnode(head, num):
    ptr = head

    while ptr != head:
        if ptr.num == num:
            return ptr
        ptr = ptr.next
    return ptr

def insertnode(head,after,num,salary,name):
    InsertNode=Employee()
    CurNode=None
    InsertNode.num=num
    InsertNode.salary=salary
    InsertNode.name=name
    InsertNode.next=None
    if InsertNode==None:
        print("記憶體配置失敗")
        return None
    else:
        if head==None:
            head=InsertNode
            InsertNode.next=head
            return head
        else:
            if after.next==head:
                InsertNode.next=head
                CurNode=head
                while CurNode.next!=head:
                    CurNode=CurNode.next
                CurNode.next=InsertNode
                head=InsertNode
                return head
            else:
                InsertNode.next=after.next
                after.next=InsertNode
                return head

position = 0
data = [[1001, 32367], [1002, 24388], [2003, 27556], [1007, 312991],
        [1012, 42660], [1014, 25676], [2018, 44145], [2043, 52182], [1031, 32769], [1037, 21100], [1041, 32196], [1046, 25776]]
namedata = ['Allen', 'scott', 'Marry', 'John', 'Mark',
            'Ricky', 'Lisa', 'Jasica', 'Hanson', 'Amy ', 'Bob', 'Jack']
print('員工編號 薪水 員工編號 薪水 員工編號 薪水 員工編號 薪水')
print("=================================================================")
for i in range(3):
    for j in range(4):
        print('[%2d] [$%3d]' % (data[j*3+i][0], data[j*3+i][1]), end='')
        print()

head = Employee()
if not head:
    print('Error!! 記憶體配置失敗!!\n')
    sys.exit(1)

head.num=data[0][0]
head.name=namedata[0]
head.salary=data[0][1]
head.next=None
ptr=head
for i in range(1,12):
    newnode=Employee()
    newnode.num=data[i][0]
    newnode.name=namedata[i]
    newnode.salary=data[i][1]
    newnode.next=None
    ptr.next=newnode
    ptr=newnode
newnode.next=head

while True:
    print('請輸入插入其後的員工編號,如輸入的編號不再此串列中,')
    position=int(input("新輸入的員工傑典將是為此串列的第一個節點,要結束插入過程請輸入-1:"))
    if position==-1:
        break
    else:
        ptr=findnode(head,position)
        new_num=int(input("請輸入新插入的員工編號:"))
        new_salary=int(input("請輸入新插入的員工薪水"))
        new_name=input("請輸入新插入的員工姓名:")
        head=insertnode(head,ptr,new_num,new_salary,new_name)

ptr=head
print('\t員工編號 姓名\t薪水')
print("==================================")

while True:
    print('\t[%2d]\t[  %-10s]\t[%3d]' %(ptr.num,ptr.name,ptr.salary))
    ptr=ptr.next
   # print(ptr.num, ptr.name, ptr.salary)
    #print(type(ptr.num), type(ptr.name), type(ptr.salary))
    if head==ptr or head==head.next:
        break
```
